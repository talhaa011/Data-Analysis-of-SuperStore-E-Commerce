## Analysis on the basis of Order Priority

```python
t1['Order Priority'].unique()
```

```python
t1['Order Priority']=t1['Order Priority'].replace("Critical ","Critical")
```

```python
plt.figure(figsize=(8,6))
count= sns.countplot(data = t1, x = "Order Priority"  , hue = "Order Priority", palette={'Not Specified':'#944121', 'High':'#29c4a0', 'Medium':'#1d4f85', 'Low':'#83d419', 'Critical' :'#f0bc13'})
count.bar_label(count.containers[0])
count.bar_label(count.containers[1])
count.bar_label(count.containers[2])
count.bar_label(count.containers[3])
count.bar_label(count.containers[4])
plt.title("Analysis on the basis of Order Priority", fontsize=20, fontweight='bold', pad=20)
plt.savefig("Analysis on the basis of Order Priority.jpg")
plt.show()
```
